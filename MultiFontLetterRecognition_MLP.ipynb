```python
# Pack per-font dictionaries
font1 = {'A':A1, 'B':B1, 'C':C1, 'D':D1, 'E':E1, 'J':J1, 'K':K1}
font2 = {'A':A2, 'B':B2, 'C':C2, 'D':D2, 'E':E2, 'J':J2, 'K':K2}
font3 = {'A':A3, 'B':B3, 'C':C3, 'D':D3, 'E':E3, 'J':J3, 'K':K3}

fonts = [font1, font2, font3]

# Build training data: 3 fonts × 7 letters = 21 samples
X_list = []
y_list = []

for class_idx, letter in enumerate(letters):      # class_idx: 0..6
    for f in fonts:
        X_list.append(f[letter])
        y_list.append(class_idx)

X = np.vstack(X_list).astype(float)
y = np.array(y_list)

print("X shape:", X.shape)  # should be (21, 63)
print("y:", y)
```

```python
multi_clf = MLPClassifier(
    hidden_layer_sizes=(12,),   # 1 hidden layer, 12 neurons
    activation='tanh',
    solver='sgd',
    learning_rate_init=0.1,
    max_iter=3000,
    random_state=0
)

multi_clf.fit(X, y)
y_pred = multi_clf.predict(X)
print("Multi-class train accuracy:", accuracy_score(y, y_pred))
print("Predicted classes:", y_pred)
```

```python
y_binary = np.array([1 if letters[c] == 'A' else 0 for c in y])

binary_clf = MLPClassifier(
    hidden_layer_sizes=(10,),
    activation='tanh',
    solver='sgd',
    learning_rate_init=0.1,
    max_iter=3000,
    random_state=0
)

binary_clf.fit(X, y_binary)
y_bin_pred = binary_clf.predict(X)
print("Binary train accuracy:", accuracy_score(y_binary, y_bin_pred))
print("Binary predictions:", y_bin_pred)
```

```python
def add_noise_flip_1_to_minus1(x, prob=0.6):
    x_noisy = x.copy()
    for i in range(len(x_noisy)):
        if x_noisy[i] == 1 and random.random() < prob:
            x_noisy[i] = -1
    return x_noisy

def add_noise_flip_minus1_to_1(x, prob=0.3):
    x_noisy = x.copy()
    for i in range(len(x_noisy)):
        if x_noisy[i] == -1 and random.random() < prob:
            x_noisy[i] = 1
    return x_noisy

def add_noise_to_zero(x, prob=0.5):
    x_noisy = x.copy()
    for i in range(len(x_noisy)):
        if random.random() < prob:
            x_noisy[i] = 0
    return x_noisy
```

```python
def eval_with_noise(model, X_clean, y_true, noise_func, prob=0.1):
    X_noisy = np.vstack([noise_func(x, prob=prob) for x in X_clean])
    y_pred = model.predict(X_noisy)
    return accuracy_score(y_true, y_pred)

noise_prob = 0.15  # 15% noise

print("=== Multi-class (7 letters) ===")
acc_1 = eval_with_noise(multi_clf, X, y, add_noise_flip_1_to_minus1, prob=noise_prob)
acc_2 = eval_with_noise(multi_clf, X, y, add_noise_flip_minus1_to_1, prob=noise_prob)
acc_3 = eval_with_noise(multi_clf, X, y, add_noise_to_zero, prob=noise_prob)

print("Noise type 1 (1→-1) accuracy:", acc_1)
print("Noise type 2 (-1→1) accuracy:", acc_2)
print("Noise type 3 (±1→0) accuracy:", acc_3)

print("\n=== Binary (A vs not-A) ===")
accb_1 = eval_with_noise(binary_clf, X, y_binary, add_noise_flip_1_to_minus1, prob=noise_prob)
accb_2 = eval_with_noise(binary_clf, X, y_binary, add_noise_flip_minus1_to_1, prob=noise_prob)
accb_3 = eval_with_noise(binary_clf, X, y_binary, add_noise_to_zero, prob=noise_prob)

print("Noise type 1 (1→-1) accuracy:", accb_1)
print("Noise type 2 (-1→1) accuracy:", accb_2)
print("Noise type 3 (±1→0) accuracy:", accb_3)
```

```python
import os
import math
import numpy as np
import matplotlib.pyplot as plt

os.makedirs("results", exist_ok=True)

def plot_letter_grid(samples_dict, title="Letter Patterns", save_path=None):
    letters = list(samples_dict.keys())
    n = len(letters)
    cols = 4
    rows = math.ceil(n / cols)

    plt.figure(figsize=(10, 7))
    for i, letter in enumerate(letters, 1):
        arr = np.array(samples_dict[letter]).reshape(9, 7)
        plt.subplot(rows, cols, i)
        plt.imshow(arr, cmap="gray", vmin=-1, vmax=1)
        plt.title(letter)
        plt.axis("off")

    plt.suptitle(title)
    plt.tight_layout()

    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

# Example: visualize font 1
plot_letter_grid(font1, title="Font 1 Letter Patterns", save_path="results/sample_letters.png")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

os.makedirs("results", exist_ok=True)

def show_noisy_examples(letter_name="A", font_dict=font1, prob=0.15, save_path=None):
    clean = np.array(font_dict[letter_name])

    noisy1 = add_noise_flip_1_to_minus1(clean, prob=prob)
    noisy2 = add_noise_flip_minus1_to_1(clean, prob=prob)
    noisy3 = add_noise_to_zero(clean, prob=prob)

    images = [
        ("Clean", clean),
        ("1→-1", noisy1),
        ("-1→1", noisy2),
        ("±1→0", noisy3),
    ]

    plt.figure(figsize=(10, 3))
    for i, (title, img) in enumerate(images, 1):
        plt.subplot(1, 4, i)
        plt.imshow(img.reshape(9, 7), cmap="gray", vmin=-1, vmax=1)
        plt.title(title)
        plt.axis("off")

    plt.tight_layout()
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches="tight")
    plt.show()
    plt.close()

show_noisy_examples(letter_name="A", font_dict=font1, prob=0.15, save_path="results/noisy_examples.png")
```
